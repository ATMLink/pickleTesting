```python
# Cell 2: 比对三版本输出

import json, difflib
from itertools import combinations

files = [
    "pickle_p4_py3.7.json",
    "pickle_p4_py3.8.json",
    "pickle_p4_py3.9.json",
]

data = {fn: json.load(open(fn, encoding='utf-8')) for fn in files}

for name in data[files[0]]:
    print(f"\n=== {name} ===")
    for a, b in combinations(files, 2):
        h1 = data[a][name].split()
        h2 = data[b][name].split()
        diff = list(difflib.unified_diff(
            h1, h2,
            fromfile=a, tofile=b,
            lineterm=''
        ))
        if diff:
            print(f"\n-- Diff between {a} and {b}:")
            for line in diff:
                print(line)
        else:
            print(f"-- No diff between {a} and {b}")
```
